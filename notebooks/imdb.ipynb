```python
##Import Library yang digunakan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH_DATA = '../mydata/'
```

```python
#Menampilkan Data yang akan digunakan
df = pd.read_csv(PATH_DATA+'imdb.csv')
```

```python
##Menghapus kolom yang tidak diperlukan
df.drop(['cast','director','composer'],axis = 1,inplace = True)
df
```

```python
##Mengecek missing value pada data
df.isnull().sum()
```

```python
##Cleaning Data/Menghapus bagian yang mengandung (NaN data)
df.dropna(subset = ['kind', 'country', 'rating', 'genre'], inplace = True)
df.reset_index(inplace = True)
df.isnull().sum()
```

```python
df_fill = df.fillna("['English']")
df_fill.head()
```

```python
##Visualisasi jumlah dan jenis tayangan melalui plot 'bar'
plt.figure(figsize=(13,8))
plot = sns.countplot(df['kind'],palette = ['pink','blue','yellow','lightblue','green','red','coral','grey']);
plt.title('Jumlah tiap jenis tayangan')
plt.xlabel('Kind')
plt.ylabel('Jumlah')
plt.show()
```

```python
##Mengerucutkan jenis tayangan ke dalam jenis yang serupa dan mengoutputkan jumlahnya
kind_new = pd.DataFrame(columns = ['kind', 'count']) 
unique_kind = df['kind'].unique()
count_kind = df['kind'].value_counts()

new_datas = {
    'movie' : 0,
    'series' : 0,
    'episode' : 0,
    'video game' : 0,
    'tv short' : 0,
}

for idx, kind in enumerate(unique_kind) : 
    if 'movie' in kind : 
        new_datas['movie'] += count_kind.loc[kind]
    elif 'series' in kind :
        new_datas['series'] += count_kind.loc[kind]
    else : 
        new_datas[kind] += count_kind.loc[kind]
for key, values in new_datas.items() : 
    kind_new = kind_new.append({'kind' : key, 'count' : values}, ignore_index = True)

kind_new
```

```python
##Visualisasi data jenis tayangan yang telah dikerucutkan melalui 'pie chart'
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
kind = ['movie','series','episode','video game','tv short']
count = [7340,973,299,18,13]
ax.pie(count, labels = kind, autopct = '%1.2f%%')
plt.show()
```

```python
##Mengecek berbgai genre yang terdapat pada data
df["genre"].head(60)
```

```python
##Menampilkan genre berbagai tayangan yang diindikasikan dengan True atau False sesuai genre judulnya
genre_list = ['Documentary','Animation','Family','Crime','Horror','Action','Sci-Fi','Comedy','Biography','Thriller','Romance','Mystery','Drama','Adventure']

for genre in genre_list :
    df[genre] = df['genre'].apply(lambda x : True if x.find(genre) != -1 else False)

df.head(3)
```

```python
##Menampilkan jumlah total judul tayangan berdasarkan genre
df_genre = pd.DataFrame(columns = ['genre', 'count'])
for genre in genre_list : 

    genre_data = {
        'genre' : genre,
        'count' : df[genre].value_counts().iloc[1]
    }
    df_genre = df_genre.append(genre_data, ignore_index = True)
df_genre.head(14)
```

```python
##Visualisasi jumlah total judul pada tiap genre
plt.figure(figsize = (15, 10))
plt.title('Number of genre', fontsize = 10)
plot = sns.barplot(data = df_genre, x = 'count', y = 'genre')
```

```python
##Menghitung jumlah dari variable countries
df['countries'] = df['country'].apply(lambda x : True if x.find(',') != -1 else False)
df['countries'].value_counts()
```

```python
##Menampilkan total jumlah tayangan yang dihasilkan oleh tiap negara
country_count = df.where(df['countries'] == False)['country'].value_counts().to_frame()
country_count.reset_index(inplace = True)
country_count.rename(columns = {'index' : 'country', 'country' : 'count'}, inplace = True)
country_count.head(15)
```

```python
##Visualisasi banyak jumlah tayangan yang dihasilkan oleh negara melalui plot 'bar'
plt.figure(figsize = (25, 30))
plt.title('Number of content by Country', fontsize = 15)
sns.barplot(data = country_count, x = 'count', y = 'country')
```

```python
##Melihat negara dengan total judul dirilis terbanyak sepanjang waktu
top_countries = country_count[0:7]
top_countries
```

```python
##Visualisasi berupa 'pie chart' untuk data 7 negara teratas yang memproduksi judul terbanyak di netflix
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
country = ['United States','United Kingdom','Japan','India','Canada','Hongkong','France']
number = [3972,788,588,265,220,214,169]
ax.pie(number, labels = country, autopct = '%1.2f%%')
plt.show()
```

```python
##Menghitung rata rating tiap negara
country_df = df.where(df['countries'] == False).groupby('country')['rating'].mean().sort_values(ascending = False).to_frame()
country_df.reset_index(inplace = True)
country_df.head(10)
```

```python
##Visualisasi data mean rating tiap negara
plt.figure(figsize = (20, 25))
plt.title("Means of Rating by Country", fontsize = 18)
sns.barplot(data = country_df, x = "rating", y = "country")
```
